```python
import pandas as pd
import numpy as np
from collections import Counter

pantheria_data = pd.read_csv('pantheria_download.csv')
iucn_data = pd.read_csv('iucn_download.csv')
pantheria_data.groupby("pred.name").count()
```

```python
iucn_data.drop(columns=["t.eol_pk", "t.resource_ok", "t.source", "pred.uri", "obj.uri", "obj.name"], inplace=True)
```

```python
def column_feature(data, name, pred_name, target_column = "measurement"):
    '''
    Returns a filtered version of "data", which is the result of a query from the EOL database. 
    Inputs: 
    data: A Dataframe
    name: The name to assign the data column from the dataframe
    pred_name: The predicate (data classification) to filter on
    target_column: 'measurement', 't literal' or 'lit name', depending on the column that holds the relevant information.
    
    '''
    #What column do we want?
    if target_column == "measurement":
        data = data.rename({"t.normal_measurement":name}, axis=1)[["p.page_id", name, "units.uri", "units.name", "pred.name"]][data["pred.name"]==pred_name]
        data[name] = data[name].replace({None: -10000}).astype(float)
        return data
    elif target_column == "t literal":
        return data.rename({"t.literal":name}, axis=1)[["p.page_id", name, "pred.name"]][data["pred.name"]==pred_name]
    elif target_column == "lit name": 
        return data.rename({"lit.name":name}, axis=1)[["p.page_id", name, "pred.name"]][data["pred.name"]==pred_name]
    return None

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
# Helper function for printing confusion matrices (see: https://gist.github.com/shaypal5/94c53d765083101efc0240d776a23823)

# prints confusion matrix as a heatmap which is nicer to visaulize

def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=18):
    """Prints a confusion matrix, as returned by sklearn.metrics.confusion_matrix, as a heatmap.
    
    Arguments
    ---------
    confusion_matrix: numpy.ndarray
        The numpy.ndarray object returned from a call to sklearn.metrics.confusion_matrix. 
        Similarly constructed ndarrays can also be used.
    class_names: list
        An ordered list of class names, in the order they index the given confusion matrix.
    figsize: tuple
        A 2-long tuple, the first value determining the horizontal size of the ouputted figure,
        the second determining the vertical size. Defaults to (10,7).
    fontsize: int
        Font size for axes labels. Defaults to 14.
        
    Returns
    -------
    matplotlib.figure.Figure
        The resulting confusion matrix figure
    """
    df_cm = pd.DataFrame(confusion_matrix, index=class_names, columns=class_names, )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d", annot_kws = {'fontsize': fontsize})
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('True label', fontsize=fontsize)
    plt.xlabel('Predicted label', fontsize=fontsize)
    return fig
```

```python
habitat_iucn = column_feature(iucn_data, "IUCN habitat", "habitat", "lit name")
conservation_status = column_feature(iucn_data, "Conservation Status", "conservation status", "lit name")


body_mass = column_feature(pantheria_data, "Body Mass", "body mass", "measurement").groupby("p.page_id").max().reset_index()
trophic_level = column_feature(pantheria_data, "Trophic level", "trophic level", "lit name").groupby("p.page_id").max().reset_index()
human_pop_density = column_feature(pantheria_data, "Human population density", "human population density", "measurement").groupby("p.page_id").max().reset_index()
geog_range = column_feature(pantheria_data, "Geographic Range", "geographic range (size of area)", "measurement").groupby("p.page_id").max().reset_index()
```

```python
pop_features = trophic_level.merge(human_pop_density, on="p.page_id", how="outer").merge(geog_range, on="p.page_id", how="outer").merge(body_mass, on="p.page_id", how="outer").merge(habitat_iucn, on='p.page_id')
```

```python
target_and_features = conservation_status.merge(pop_features, on="p.page_id", how='inner')
target = target_and_features['Conservation Status']
```

```python
features = target_and_features[["Human population density", "Geographic Range", "Body Mass"]].merge(pd.get_dummies(target_and_features["Trophic level"]), left_index=True, right_index=True)
```

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
frame = pd.DataFrame(features_scaled).replace({None: "-10000"})
```

```python
target.replace({"extinct in the wild" : 'extinct', 'conservation dependent' : 'critically endangered'}, inplace=True)
drop_unknown = ((target != 'data deficient'))
drop_unknown_trophic = !(features_scaled["Trophic level"] == None)
```

```python
#names = ['extinct','critically endangered','endangered', 'vulnerable','near threatened','least concern']
names = ["Endangered", "Not endangered"]
#names = ["Endangered", "Unclear", "Not endangered"]

target_scaler = {}
#target_scaler = {'extinct':0, 'critically endangered': 1, 'endangered': 2, 'vulnerable': 3, 'near threatened': 4, 'least concern': 5}
target_scaler = {'extinct':0, 'critically endangered': 0, 'endangered': 0, 'vulnerable': 2, 'near threatened': 2, 'least concern': 2}
target_num = target.replace(target_scaler)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(frame[drop_unknown], target_num[drop_unknown], test_size=0.4, stratify=target_num[drop_unknown])
X_test, X_validate, y_test, y_validate = train_test_split(X_test, y_test, test_size=0.5, stratify=y_test)

y_train = y_train.astype('string')
y_test = y_test.astype('string')
y_validate = y_validate.astype('string')

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_over, y_over = ros.fit_resample(X_train,y_train)

from imblearn.over_sampling import ADASYN
ada = ADASYN()
X_resample, y_resample = ada.fit_resample(X_train, y_train)
y_resample = y_resample.astype('string')
```
